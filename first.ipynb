```python
money = 1000

if money > 1500:
    print('커피를 마셔요')
    print('빵도 먹어요')
elif money > 500:
    print('사탕')
else:
    print('물을 마셔요')
print('조건문 종료')
```

```python
# 조건문 활용 : 짝수 홀수 판단
# 2 -> 짝수
# 3 -> 홀수

num_input = int(input("숫자를 입력하세요:"))

if num_input % 2 == 0:
    print("짝수")
else:
    print("홀수")
```

```python
# 성적처리 프로그램
# 90점 이상이면 'A'
# 80점 이상 90점 미만이면 'B'
# 70점 이상 80점 미만이면 'C'
# 60점 이상 70점 미만이면 'D'
# 아니면 'F'

score = int(input("점수를 입력하세요:"))

if score >= 90:
    result = 'A'
elif score >= 80 and score < 90:
    result = 'B'
elif score >= 70 and score < 80:
    result = 'C'
elif score >= 60 and score < 70:
    result = 'D'
else:
    result = 'F'

print(f'학점은 {result}입니다')
```

```python
# 수학점수, 영어점수
# 수학점수와 영어점수가 둘다 90점 이상이면 '참 잘했어요'
# 수학점수만 90점 이상이면 '영어를 좀 더 노력하세요'
# 영어점수만 90점 이상이면 '수학를 좀 더 노력하세요'
# 그렇지 않으면 '많이 노력하세요'

math_score = 80
eng_score = 50

if math_score >= 90 and eng_score >=90:
    print('참 잘했어요')
elif math_score >= 90:
    print('영어를 좀 더 노력하세요')
elif eng_score >= 90:
    print('수학를 좀 더 노력하세요')
else:
    print('많이 노력하세요')


if math_score >= 90:
    if eng_score >= 90:
        print('참 잘했어요')
    else:
        print('좀 더 노력하세요')
else:
    print('많이 노력하세요')

```

```python
# for 반복문과 range() 함수를 사용해서 1부터 10까지의 합을 구해주세요!
result_add = 1
result_mul = 1

for i in range(2,11):
    result_add += i
    result_mul *= i

print(f'합은 {result_add}, 곱은 {result_mul} 입니다.')
```

```python
# for 반복문과 range() 함수를 사용해서 1부터 10까지의 곱을 구해주세요!
result = 1

for i in range(1,11):
    result *= i

print(f'합은 {result} 입니다.')
```

```python
# for 반복문과 range() 함수를 사용해서 1부터 10까지의 합을 구해주세요!
my_num = 100

sum = 0
for i in range(1, my_num + 1):
    sum+=i

print(sum)
```

```python
# while 반복문을 이용해서 1부터 10까지 더하되 합이 20이 넘는 순간의 정수값을 출력하는 코드 작성
i = 0
sum = 0
while i <= 10:
    i += 1
    sum += i
    if sum >= 20:
        print(f'정수:{i}, 합 :{sum}')
        break


```

```python
# 숫자 맞추기 게임
import random 
com = random.randint(1,100)

# print(com)
while True:
    user = int(input('원하는 숫자 입력 : '))

    if com == user:
        print("정답입니다.")
        break
    elif com > user:
        print("작다")
    else:
        print("크다")

```
